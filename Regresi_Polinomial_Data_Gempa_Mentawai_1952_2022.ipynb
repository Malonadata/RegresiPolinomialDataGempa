```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("https://www.dropbox.com/s/nmn0ggd1m0el7xx/Data%20Gempa%201952-2022.csv?dl=1",sep=";")

dataset
```

```python
y = dataset.loc[:, ['mag']]
y
```

```python
X = dataset.loc[:, ['depth']]

X
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
from matplotlib.pyplot import figure
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Visualizing the Linear Regression results
def viz_linear():
    figure(figsize=(10, 5))
    plt.scatter(X, y, color='red')
    plt.plot(X, lin_reg.predict(X), color='blue')
    plt.title('Kedalaman Vs Magnitudo (Linear Regression)')
    plt.xlabel('Kedalaman')
    plt.ylabel('Magnitudo')
    plt.grid()
    plt.show()
    return
viz_linear()
```

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly2_reg = PolynomialFeatures(degree=2)
X_poly2 = poly2_reg.fit_transform(X)
pol2_reg = LinearRegression()
pol2_reg.fit(X_poly2, y)

# Visualizing the Polymonial Regression results
def viz_polymonial():
    figure(figsize=(10, 5))
    plt.scatter(X, y, color='red')
    plt.plot(X, pol2_reg.predict(poly2_reg.fit_transform(X)), color='blue')
    plt.title('Kedalaman Vs Magnitudo (Polynomial Regression Orde 2)')
    plt.xlabel('Kedalaman')
    plt.ylabel('Magnitudo')
    plt.show()
    return
viz_polymonial()

# Additional feature
# Making the plot line (Blue one) more smooth
def viz_polymonial_smooth():
    Xgrid = np.arange(0.3, 153.2, 0.1)
    Xgrid = Xgrid.reshape(len(Xgrid), 1) 
    # Visualizing the Polymonial Regression results
    figure(figsize=(10, 5))
    plt.scatter(X, y, color='red')
    plt.plot(Xgrid, pol2_reg.predict(poly2_reg.fit_transform(Xgrid)), color='blue')
    plt.title('Kedalaman Vs Magnitudo (Polynomial Regression Orde 2)')
    plt.xlabel('Kedalaman')
    plt.ylabel('Magnitudo')
    plt.grid()
    plt.show()
    return
viz_polymonial_smooth()
```
